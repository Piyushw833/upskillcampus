```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datafile.csv')
df
```

```python
col = df.columns
```

```python
col
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
import random
color = ['green','red','blue','yellow','orange']
for i in range (len(df['Crop'])):  
    plt.plot(col[1:],df.iloc[i].values[1:] , color=random.choice(color))
    plt.xticks(rotation = 60)
    plt.title(df['Crop'][i])
    plt.xlabel('Year')
    plt.ylabel('Production')
    plt.grid()
    plt.show()
```
